```python
import numpy as np
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore")
```

```python
f = lambda x: (3 * x**2) + 4

xs = np.arange(-1.5, 1, 0.05)
ys = f(xs)
ys
```

```python
IDX=11
plt.plot(xs, ys)
plt.plot(xs[IDX], ys[IDX], "ro")
plt.show()
```

What is the slope at this point?

![Image](https://wikimedia.org/api/rest_v1/media/math/render/svg/aae79a56cdcbc44af1612a50f06169b07f02cbf3)

A function of **a** real variable **f(x)** is differentiable at a point a of its domain,\
if its domain contains an open interval I containing a, and the limit exists.

*ref: wikipedia*

```python
h = 0.0001
l = (f(xs[IDX] + h) - f(xs[IDX])) / h
l
```

```python
ls = (f(xs + h) - f(xs)) / h
ls
```

```python
IDX=11
plt.plot(xs, ls)
plt.plot(xs[IDX], ls[IDX], "ro")
plt.show()
```

```python
f = lambda x: (3 * x**3) + 4
xs = np.arange(-1.5, 1, 0.05)
ys = f(xs)

IDX=11
plt.plot(xs, ys)
plt.plot(xs[IDX], ys[IDX], "ro")
plt.show()

h = 0.0001
ls = (f(xs + h) - f(xs)) / h
ls

plt.plot(xs, ls)
plt.plot(xs[IDX], ls[IDX], "ro")
plt.show()
```
