```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
square = lambda x: x**2
sigmoid = lambda x:  1 / (1 + np.exp(-x))

x = 2
square(x), sigmoid(x), square(sigmoid(x)), sigmoid(square(x))
```

```python
xs = np.arange(-3, 3, 0.01)
xs
```

```python
# square(x)
sq = square(xs)
plt.title("square(x)")
plt.plot(xs, sq)
plt.show()
```

```python
# sigmoid(x)
si = sigmoid(xs)
plt.title("sigmoid(x)")
plt.plot(xs, si)
plt.show()
```

```python
# sigmoid(square(x))
sisq = sigmoid(square(xs))
plt.title("sigmoid(square(x))")
plt.plot(xs, sisq)
plt.show()
```

```python
# square(sigmoid(x))
sqsi = square(sigmoid(xs))
plt.title("square(sigmoid(x))")
plt.plot(xs, sqsi)
plt.show()
```

```python
# derivative - square(sigmoid(x))
h = 0.0001
dsi = (sigmoid(square(xs) + h) - sigmoid(square(xs))) / h
dsq = (square(xs + h) - square(xs)) / h
dsidsq = dsi * dsq
plt.title("derivative - square(sigmoid(x))")
plt.plot(xs, dsidsq, c="r")
plt.show()
```

```python
# derivative - sigmoid(square(x))
h = 0.0001
dsq = (square(sigmoid(xs) + h) - square(sigmoid(xs))) / h
dsi = (sigmoid(xs + h) - sigmoid(xs)) / h
dsqdsi = dsi * dsq
plt.title("derivative - sigmoid(square(x))")
plt.plot(xs, dsqdsi, c="g")
plt.show()
```
